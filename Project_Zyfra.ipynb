# Calcular la recuperación de oro en el conjunto de entrenamiento y compararla con los valores reales para verificar si los cálculos son correctos.

$$ Recuperación =  \frac{C*(F - T)}{F*(C - T)}\,*100\% $$

dónde:

* C: la proporción de oro en el concentrado justo después de la flotación (para saber la recuperación del concentrado rougher)/después de la purificación (para saber la recuperación del concentrado final).

* F: la proporción de oro en la alimentación antes de la flotación (para saber la recuperación del concentrado rougher)/en el concentrado justo después de la flotación (para saber la recuperación del concentrado final).

* T: la proporción de oro en las colas rougher justo después de la flotación (para saber la recuperación del concentrado rougher)/después de la purificación (para saber la recuperación del concentrado final).

```python
# Calcula la recuperación de la característica rougher.output.recovery

df_train['rougher.output.recovery']
```

```python
# C : rougher.output.concentrate_au
# F : rougher.input.feed_au
# T: rougher.output.tail_au


def recovery_calculation(df, stage = 'rougher'):
    if stage == 'rougher':
        C =  df['rougher.output.concentrate_au']
        F =  df['rougher.input.feed_au']
        T =  df['rougher.output.tail_au']
    
    elif stage == 'final':
        C = df['final.output.concentrate_au']
        F = df['rougher.output.concentrate_au']
        T = df['final.output.tail_au']
    
    else:
        print('Indicar etapa adecuada (rougher o final)')
        return np.nan    
    
    numerador = C*(F-T)
    denominador = F*(C-T)
    recovery = 100*(numerador/denominador)
    
    return recovery
```

```python
rougher_data = df_train[['rougher.output.concentrate_au',
                         'rougher.input.feed_au',
                         'rougher.output.tail_au', 
                         'rougher.output.recovery']].dropna()
rougher_recovery = recovery_calculation(rougher_data, 'rougher')
```

```python
print(rougher_recovery)
rougher_data.shape
```

```python
# Calcular el Error Absoluto Medio (MAE)

mae = mean_absolute_error(rougher_recovery, rougher_data['rougher.output.recovery'])
print(f"MAE: {mae}")
```

El Error Absoluto Medio (MAE) = 9.303415616264301e-15 sugiere que el cálculo de la recuperación de oro es correcto. 

Con un MAE tan bajo (prácticamente cero), podemos estar seguros de que la recuperación se está calculando correctamente.

```python
# Definir las columnas de interés relacionadas con las concentraciones en cada etapa
concentration_columns = [
    'rougher.input.feed_ag', 'rougher.input.feed_pb', 'rougher.input.feed_sol', 'rougher.input.feed_au',
    'primary_cleaner.input.sulfate', 'primary_cleaner.input.depressant', 'primary_cleaner.input.xanthate',
    'final.output.recovery'
]

# Verificar si hay valores anómalos con diagramas de caja
plt.figure(figsize=(12, 6))
df_train[concentration_columns].boxplot(rot=45)
plt.title("Distribución de concentraciones en diferentes etapas")
plt.show()
```

```python
# Calcular límites usando el rango intercuartílico (IQR)
def detect_outliers(df, columns):
    outliers = pd.DataFrame()
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outliers[col] = (df[col] < lower_bound) | (df[col] > upper_bound)
    return outliers

outlier_flags = detect_outliers(df_train, concentration_columns)

# Contar valores anómalos por columna
outlier_counts = outlier_flags.sum()
print("Cantidad de valores anómalos por columna:")
print(outlier_counts)

# Eliminar filas con valores anómalos
df_train_cleaned= df_train[~outlier_flags.any(axis=1)]

# Mostrar cuántas filas se eliminaron
print(f"Filas originales: {len(df_train)}, Filas después de eliminar anomalías: {len(df_train_cleaned)}")
```

# Modelo de ML

Las features deben ser las columnas del dataset de test

```python
# Definir objetivo

target = ['rougher.output.recovery', 'final.output.recovery']
```

```python
y_train = df_train_cleaned[target].reset_index(drop = True)
X_train = df_train_cleaned.drop(target, axis = 1)
X_train = X_train.drop(['date'], axis = 1)
```

```python
def compute_smape(y, y_pred):
    n = len(y)
    real = abs(y)
    pred = abs(y_pred)
    diff = abs(y - y_pred)
    smape =  (1/n)*np.sum(diff /((real + pred)/2))*100
    return smape

def smape_final(y, y_pred):
    
    y_rougher = y.iloc[:,0]
    y_pred_rougher = y_pred[:,0]
    
    y_final = y.iloc[:,1]
    y_pred_final = y_pred[:,1]
    
    smape_rougher = compute_smape(y_rougher, y_pred_rougher)
    smape_final = compute_smape(y_final, y_pred_final)
    
    return (0.25*smape_rougher + 0.75*smape_final)

smape_scorer = make_scorer(smape_final)
```

```python
# Modelo Regresión Lineal

lr_model = LinearRegression()

lr_score = cross_val_score(lr_model, X_train, y_train, scoring=smape_scorer, cv = 5)

lr_final_score = lr_score.mean()

print('Puntajes sMAPE para cada iteración:', lr_score)
print('Modelo de Regresión Lineal | sMAPE = {:.6f}'.format(lr_final_score))
```

```python
# Modelo Random Forest Regressor.

rf_model = RandomForestRegressor(n_estimators=100, random_state=12345, max_depth = 2)

rf_score = cross_val_score(rf_model, X_train, y_train, scoring=smape_scorer, cv=5)
rf_final_score = rf_score.mean()

print('Puntajes sMAPE para cada iteración:', rf_score)
print('Modelo Random Forest | sMAPE = {:.6f}'.format(rf_final_score))
```

# Probar el model Random Forest en el conjunto de test. 

```python
# Preparar datos para el modelo
target = ['rougher.output.recovery', 'final.output.recovery']
y_train = df_train_cleaned[target].reset_index(drop=True)
X_train = df_train_cleaned.drop(target, axis=1).drop(['date'], axis=1)
X_test = df_test.drop(target, axis=1).drop(['date'], axis=1)
y_test = df_test[target].reset_index(drop=True)
```

```python
def smape_final(y, y_pred):
    y_rougher = y.iloc[:, 0]
    y_pred_rougher = y_pred[:, 0]
    y_final = y.iloc[:, 1]
    y_pred_final = y_pred[:, 1]
    smape_rougher = compute_smape(y_rougher, y_pred_rougher)
    smape_final = compute_smape(y_final, y_pred_final)
    return (0.25 * smape_rougher + 0.75 * smape_final)

smape_scorer = make_scorer(smape_final)

# Entrenar el modelo Random Forest
rf_model = RandomForestRegressor(n_estimators=100, random_state=12345, max_depth = 2)
rf_model.fit(X_train, y_train)

# Hacer predicciones en el conjunto de prueba
y_pred = rf_model.predict(X_test)

# Evaluar el modelo en la muestra de prueba
final_smape = smape_final(y_test, y_pred)

print("sMAPE en la muestra de prueba: {:.6f}".format(final_smape))
```

El Random Forest Regressor obtuvo un sMAPE = 9.537, mejor que la Regresión Lineal (sMAPE = 9.835), lo que indica que el modelo de Random Forest tiene una mejor capacidad de predicción para este problema.

El modelo Random Forest obtuvo un sMAPE menor en la prueba (9.532) en comparación con la validación cruzada (9.537), lo que sugiere que generaliza bien.
